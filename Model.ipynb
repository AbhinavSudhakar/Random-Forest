```python
data["Motion"]=data["Motion"].map({"Sitting":0,"Sit and Stand":1,"Walking":2})
```

```python
data.columns
```

```python
for i in data.columns:
  print(i)
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS=MinMaxScaler()
for i in data.columns:
  if i=="Motion":
    break
  data[i]=MMS.fit_transform(data[[i]])
```

```python
data
```

```python
data.shape
```

```python
data.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

LOR=LogisticRegression()
X=data[["AS5600 - Ankle","AS5600- Knee","MPU - Acc_x - Ankle","MPU-Acc_y - Ankle","MPU - Acc_z - Ankle","MPU - Gyro_x - Ankle","MPU - Gyro_y - Ankle","MPU - Gyro_z - Ankle","MPU - Acc_x - Knee","MPU - Acc_y - Knee","MPU - Acc_z - Knee","MPU - Gyro_x - Knee","MPU - Gyro_y - Knee","MPU - Gyro_z - Knee"]]
y=data[["Motion"]]
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
y_test
```

```python
LOR.fit(X_train,y_train)
```

```python
predict=LOR.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred_RF=model.predict(X_test)
y_pred_RF
```

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
plt.figure(figsize=(12,7))
cm=confusion_matrix(y_test,y_pred_RF)
sns.heatmap(cm,annot=True)
plt.xlabel("predicted")
plt.ylabel("truth")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_RF))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_RF))
```
